Importazione del dataset e di `pandas`

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
```

Mostra il numero di righe e di colonne del dataset:

```python
df.shape
```

Mostra le prime 10 righe del dataset:

```python
df.head(10)
```

Mostra le ultime 10 righe del dataset:

```python
df.tail(10)
```

Mostra il nome (o i nomi) del passeggero/i più giovane del Titanic:

```python
print(df[df.Age == df.Age.min()]['Name'])
print(df[df.Age == df.Age.min()]['Age'])
```

Mostra il numero totale di passeggeri e dei sopravvissuti:

```python
total_passengers = df.shape[0]
num_survived = df['Survived'].sum()
print(f"Total passengers: {total_passengers}")
print(f"Number of survivors: {num_survived}")
```

Mostra la percentuale di sopravvissuti sul totale:

```python
survival_rate = (num_survived / total_passengers) * 100
print(f"Survival rate: {survival_rate:.2f}%")
```

Mostra la media di età dei sopravvissuti e dei non sopravvissuti:

```python
avg_age_survived = df[df['Survived'] == 1]['Age'].mean()
avg_age_not_survived = df[df['Survived'] == 0]['Age'].mean()
print(f"Average age of survivors: {avg_age_survived:.2f}")
print(f"Average age of non-survivors: {avg_age_not_survived:.2f}")
```

Mostra i sopravvissuti per classe sulla nave:

```python
class_survival = df[df['Survived'] == 1]['Pclass'].value_counts()
print("Survivors per class:")
print(class_survival)
```

Mostra i sopravvissuti per genere:

```python
gender_survival = df[df['Survived'] == 1]['Sex'].value_counts()
print("Survivors by gender:")
print(gender_survival)
```

Importa Matplotlib:

```python
import matplotlib.pyplot as plt
```

Crea un grafico a torta della percentuale di sopravvisuti e della percentuale di non sopravvissuti:

```python
labels = 'Survived', 'Not Survived'
sizes = [num_survived, total_passengers - num_survived]
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title("Survival distribution")
plt.show()
```

Crea un grafico a barre dei sopravvissuti per classe:

```python
classes = [1, 2, 3]
num_survivors_by_class = [class_survival[c] for c in classes]
plt.figure(figsize=(8, 6))
plt.bar(classes, num_survivors_by_class, align='center', alpha=0.7)
plt.xticks(classes, ["1st", "2nd", "3rd"])
plt.xlabel("Class")
plt.ylabel("Number of survivors")
plt.title("Survivors by class")
plt.show()
```

Crea un grafico con due curve che rappresentino rispettivamente i sopravvissuti ed i non sopravvissuti per le tre classi della nave:



```python
plt.figure(figsize=(8, 6))

avg_age_survived = [df[(df['Survived'] == 1) & (df['Pclass'] == cls)]['Age'].mean() for cls in classes]
avg_age_not_survived = [df[(df['Survived'] == 0) & (df['Pclass'] == cls)]['Age'].mean() for cls in classes]

plt.plot(classes, avg_age_survived, marker='o', color='b', label='Survived')
plt.plot(classes, avg_age_not_survived, marker='s', linestyle='--', color='r', label='Not Survived')


plt.xticks(classes, ["1st", "2nd", "3rd"])
plt.xlabel("Class")
plt.ylabel("Average Age")
plt.title("Average Age by Class and Survival")
plt.legend()

plt.show()
```
